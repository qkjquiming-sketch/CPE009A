Arithmetic Operations

```python
x = 5
y = 3

print(x + y)    # Addition
print(x - y)    # Subtraction
print(x * y)    # Multiplication
print(x / y)    # Division
print(x % y)    # Modulo
print(x ** y)   # Exponentiation
print(x // y)   # Floor division
```

Assignment Operators

```python
s = 1

s += 3
print(f"s += 3 : {s}")
s -= 3
print(f"s -= 3 : {s}")
s *= 3
print(f"s *= 3 : {s}")
s /= 3
print(f"s /= 3 : {s}")
s %= 3
print(f"s %= 3 : {s}")
s //= 3
print(f"s //= 3 : {s}")
s **= 3
print(f"s **= 3 : {s}")

s = 1

s &= 3
print(f"s &= 3 : {s}")
s |= 3
print(f"s |= 3 : {s}")
s ^= 3
print(f"s ^= 3 : {s}")
s >>= 3
print(f"s >>= 3 : {s}")
s <<= 3
print(f"s <<= 3 : {s}")
```

Comparison Operators

```python
t = 1
u = 2

print(t == u)  # Equal
print(t != u)  # Not equal
print(t > u)   # Greater than
print(t < u)   # Less than
print(t >= u)  # Greater than or equal to
print(t <= u)  # Less than or equal to
```

Logical Operators

```python
print(t > 0 and t < 10)         # Prints TRUE if BOTH statements are true
print(t < 5 or t > 4)           # Prints TRUE if ONE of the statements is true
print(not(t > 5 and t < 10))    # Reverse the result
```

Identity Operators

```python
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x

print(x is z)
print(x is y)
print("\n")
print(x is not z)
print(x is not y)
```

Bitwise Operator

```python
v = 5
w = 3

print(v & w)  # AND Operator
print(v | w)  # OR Operator
print(v ^ w)  # XOR Operator
print(~v)     # NOT Operator
print(v << 1) # Left Shift Operator
print(v >> 1) # Right Shift Operator
```
